```python
import matplotlib.pyplot as plt

x1 = data["number_courses"]
y1 = data["Marks"]

x2 = data["time_study"]
y2 = data["Marks"]

plt.figure(figsize=(15, 10))  # Adjust the figure size if needed

plt.subplot(1, 2, 1)
plt.plot(x1, y1, color='blue')  # Change the color to blue
plt.title("number_courses vs.Marks")
plt.xlabel("number_courses")
plt.ylabel("Marks")

plt.subplot(1, 2, 2)
plt.plot(x2, y2, color='green')  # Change the color to green
plt.title("time_study vs. Marks")
plt.xlabel("time_study")
plt.ylabel("Marks")

plt.tight_layout()  # Adjust subplot spacing for better visualization
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
figure=px.scatter(data_frame=data,x='Marks',y="time_study",size="time_study",trendline="ols")
figure.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Model=LinearRegression()
Model.fit(features,response)
```

```python
Intercept=float(Model.intercept_)
print("The intercept of the fitted model is:=", Intercept)
```

```python
coeficient=Model.coef_
print("The Coefficient of the fitted model is:" , coeficient)
```
